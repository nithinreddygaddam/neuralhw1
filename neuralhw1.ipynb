```python
class_score = float(input("Enter the class score: "))

if class_score >= 90:
    letter_grade = 'A'
elif class_score >= 80:
    letter_grade = 'B'
elif class_score >= 70:
    letter_grade = 'C'
elif class_score >= 60:
    letter_grade = 'D'
else:
    letter_grade = 'F'

print(f"The letter grade for {class_score} is: {letter_grade}")
```
